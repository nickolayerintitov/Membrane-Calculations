# Pressure-Deflection-Strain Calculations on Membranes

For $r=0$ (center of the membrane )  
Radial and tangential stresses are equal
$N_r = N_θ = Et\frac{b_0}{4} (\frac{Pa}{Et})^{2/3} $ [Ref. A, eq.12a]  
$N_r - νN_θ = Et ϵ_r$  (ν, Poisson's ratio)  
$N_r(1 - ν) = Et ϵ_r$   at $r=0$  
$ϵ_r = (1 - ν)\frac{N_r}{Et}$  
$$ϵ_r = \frac{(1 - ν)b_0}{4} (\frac{Pa}{Et})^{2/3}$$  
$h = \frac{a}{b_0}(\frac{Pa}{Et})^{1/3}$  [A, eq.12b]  
$ϵ_r = \frac{(1 - ν)b_0^{3}}{4} (\frac{h}{a})^{2}$  

$m$, slope of the $ϵ_r$ vs $(Pa)^{2/3}$ curve:  
$$m = \frac{(1 - ν)b_0}{4} (\frac{1}{Et})^{2/3}$$  
$$Et= (\frac{(1 - ν)b_0}{4m})^\frac{3}{2}$$


$P$: Pressure  
$a$: radius of the hole  
$Et$ : 2D Elastic modulus (Young's modulus $\times$ thickness)  
$ν$ : Possion's ratio  
$b_0$ : $ν$ dependent parameter  
$h$ = vertical deflection of the membrane

```python
import matplotlib.pyplot as plt
import scipy
from scipy import interpolate
import matplotlib as mpl
from matplotlib import cm
exp=math.exp
pi=math.pi
inf=math.inf



#WSe2

#MoS2
#v0=0.25
#Et=177
#b0=1.689

Et = 340 #N/m, 2D Young's modulus A Review on Mechanics and Mechanical Properties of 2D Materials – Graphene and Beyond

v = 0.16 # Poisson's ratio for graphene
b0 = 1.668 #for graphene

[b1,b3,b5,b7,b9,b11,b13,b15,b16,b17,b19,b21] = [0]*12
b2=-b0**-2
b4=-2/3*b0**-5
b6=-13/18*b0**-8
b8=-17/18*b0**-11
b10=-37/27*b0**-14
b12=-1205/567*b0**-17
b14=-219241/63504*b0**-20
b16=-6634069/1143072*b0**-23
b18=-51523763/5143824*b0**-26
b20=-998796305/56582064*b0**-29
b=[b0,b1,b2,b3,b4,b5,b6,b7,b8,b9,b10,b11,b12,b13,b14,b15,b16,b17,b18,b19,b20,b21]

[a1,a3,a5,a7,a9,a11,a13,a15,a16,a17,a19,a21] = [0]*12
a0 = 1/b0
a2 = (1/2)*b0**-4
a4 = (5/9)*b0**-7
a6 = (55/72)*b0**-10
a8 = (7/6)*b0**-13 
a10 = (205/108)*b0**-16
a12 = (17051/5292)*b0**-19
a14 = (2864485/508032)*b0**-22
a16 = (103863265/10287648)*b0**-25
a18 = (27047983/1469664)*b0**-28
a20 = (42367613873/1244805408)*b0**-31
c=[a0,a1,a2,a3,a4,a5,a6,a7,a8,a9,a10,a11,a12,a13,a14,a15,a16,a17,a18,a19,a20,a21]

K = sum(c)**-3


def radius_curvature(x,y):
    dx = np.gradient(x)
    dy = np.gradient(y)/dx
    ddy = np.gradient(dy)/dx
    return dy , ((1 + dy**2)**1.5)/ddy

# calculate the strain from the vertical deflection of the membrane
def strain_from_h(a=1 #metre----  radius of the hole
                     ,Δh=100 #metre---- maximum vertical  deflection of the membrane 
                     ,Δh0= 0 #metre----  initial deflecion of the beam when strain is still zero; offset deflection
                     ,ν=0.16 # Poisson's ratio of MoS2, 0.2 is typical for TMDCs
                ):
    
    
    
    σ=0.25*b0*(1-ν)*K**(2/3)
    
    return 10**2*σ*((Δh-Δh0)/a)**2 # % strain

def strain_from_pressure(P= 1 #N/m^2, pressure difference between the environment and the space below the membrane
                             ,a=1 #metre----  radius of the hole
                             ,ν=0.16
                            ):
    Δh=deflection_from_pressure(P=P/101325,a=a*10**6)
    return strain_from_h(a=a,Δh=Δh*10**-6,ν=ν)

def deflection_from_pressure(P = 1, #atm, pressure difference between the environment and the space below the membrane
                             a = 1, #μm, radius of the hole
                             Et = 340 #N/m, 2D Young's modulus
                            ):
    
    return ( (P*a**4)/(K*Et) )**(1/3) #m micrometre



def deflection(P= 1, #N/m^2, pressure_environment - pressure_below_membrane
                a=1*10**-6, #m radius of the hole
                r=0, #m radial distance from the center    
                Et=340, #N/m, 2D Young's modulus
               b0=1.668
                            ):

    return integrate.quad(lambda x:P*x/(2*Nr(P,a,x,Et,b0)), r, a)[0] #m Eq. (2) of W. B. Fichter

    h_sum = 0 # deflection
    for n in range(len(b)//2):
        h_sum +=  c[2*n] * (1-(r/a)**(2*n+2))
    return h_sum * (a) * (P*a/Et)**(1/3) # Eq. (12b) of W. B. Fichter

def Nr(P, a, r, Et,b0): # radial stress
    b_sum=0
    #ekleme
    [b1,b3,b5,b7,b9,b11,b13,b15,b16,b17,b19,b21] = [0]*12
    b2=-b0**-2
    b4=-2/3*b0**-5
    b6=-13/18*b0**-8
    b8=-17/18*b0**-11
    b10=-37/27*b0**-14
    b12=-1205/567*b0**-17
    b14=-219241/63504*b0**-20
    b16=-6634069/1143072*b0**-23
    b18=-51523763/5143824*b0**-26
    b20=-998796305/56582064*b0**-29
    b=[b0,b1,b2,b3,b4,b5,b6,b7,b8,b9,b10,b11,b12,b13,b14,b15,b16,b17,b18,b19,b20,b21]
    #ekleme bitiş
    for n in range(len(b)//2):
        b_sum+=b[2*n]*(r/a)**(2*n)
    return 0.25*(Et*P**2*a**2)**(1/3)*b_sum # Eq. (12a) of W. B. Fichter

def Nt(P, a, r, Et): # tangential stress, aka Nθ
    b_sum=0
    for n in range(len(b)//2):
        b_sum+=(2*n+1)*b[2*n]*(r/a)**(2*n)
    return 0.25*(Et*P**2*a**2)**(1/3)*b_sum # Eq. (12c) of W. B. Fichter

def strain_r_t(P, a, r, Et): # radial, tangential strain
    
    N_radial=Nr(P, a, r, Et,b0)
    N_tangential=Nt(P, a, r, Et)
    
    strain_r = (N_radial-v*N_tangential)/Et
    strain_t = (N_tangential-v*N_radial)/Et
    average = (strain_r + strain_t)/2
    
    return [strain_r, average, strain_t]

def bulge_volume(P=1, a=1*10**-6, Et=312,b0=1.668): # volume of the bulge formed
    
    return integrate.quad(lambda x:2*pi*x*deflection(P,a,r=x,Et=Et,b0=b0), 0, a)[0] #m^3


def custom_to_SI(P,a,r): #Convert to SI units
        
    P=P*101325 # atm to N/m^2
    a=a*10**-6 # μm to m
    r=r*10**-6 # μm to m

    return P,a,r

def SI_to_custom(P,a,r): #Convert to custom units
        #Convert to SI units
    P=P/101325 # N/m^2 to kPa
    a=a*10**6 # m to μm
    r=r*10**6 # m to μm

    return P,a,r
```
